```python
df = df.drop_duplicates()
```

```python
df['Result'].value_counts()
```

```python
X = df.drop('Result', axis= 1)
y = df['Result']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()

lg.fit(X_train,y_train)
y_pred = lg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

print("Accuracy :", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall   :", recall_score(y_test, y_pred))
print("F1 Score :", f1_score(y_test, y_pred))

print("\nConfusion Matrix")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
base_models = [
    ('lg', LogisticRegression(max_iter=1000, random_state=42)),
    ('dt', DecisionTreeClassifier(random_state=42)),
    ('knn', KNeighborsClassifier(n_neighbors=5)),
    ('svc', SVC(probability=True, random_state=42))
]
```

```python
meta_model = LogisticRegression(max_iter=1000, random_state=42)
```

```python
from sklearn.ensemble import StackingClassifier

classifier = StackingClassifier(
    estimators= base_models,
    final_estimator = meta_model,
    cv=5
)
```

```python
classifier.fit(X_train,y_train)
```

```python
y_pred_s = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred_s)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, y_pred_s))
print(classification_report(y_test, y_pred_s))
```

```python
import joblib

joblib.dump(classifier, "stacking_model.pkl")

print("Model saved successfully!")
```
